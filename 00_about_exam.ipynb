## Dimensions and Measures
>Nouns
### Dimensions
Qualitative values *can only find the frequencies and mode*
- Date
- Names
- GeoData

### Measures
Numeric, qualitative values *values can be aggregated*
- Profit
- Quantity
- Weight

## Discrete and Continuous
>Adjectives
### Discrete `blue pill`
Finite range
- Numeric dimensions with a datatype of `string` of `boolean`

### Continuous `green pill`
Infinite range
